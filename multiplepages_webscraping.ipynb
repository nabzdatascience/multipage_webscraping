```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
ph_name=[]
ph_price=[]
pages=input("Please enter the number of pages to be scraped: ")
for i in range(1,int(pages)+1):
    url="https://www.flipkart.com/search?q=redmi+a2+mobile&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_6_3_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_6_3_na_na_na&as-pos=6&as-type=RECENT&suggestionId=redmi+a2+mobile&requestId=6974ac7e-cc0a-4cfc-98bc-8c1aa75faa8b&as-searchtext=red"
    response=requests.get(url)
    #print(response)
    soup=BeautifulSoup(response.text,"html.parser")
    for name in soup.find_all("div",class_="_4rR01T"):
        ph_name.append(name.text)
    print(ph_name)
    for price in soup.find_all("div",class_="_30jeq3 _1_WHN1"):
        ph_price.append(price.text)
    print(ph_price)  
    print(len(ph_price))
```

```python
raw_data={"phone_name":ph_name,
         "phone_price":ph_price}
```
